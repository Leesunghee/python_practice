```python
class Student():
    def __init__(self, name):
        self.name = name
        
    def study(self):
        print("공부를 합니다.")
        
    def introduce(self):
        print("저는 {name} 입니다.".format(name=self.name))
        
# class DataScienceStudent(...):
#     ...
#     # 저는 데이터 사이언스 
    
# class WebProgrammingStudent(...):
#     ...
#     # 저는 데이터 사이언스 

class WPSStudent(Student):
    def introduce(self):
        print("저는 웹프스 {name} 입니다.".format(name=self.name))
        
    def coding(self):
        print("vim 과 bash 를 이용해서 웹 개발을 합니다.")
        
class DSSStudent(Student):
    def introduce(self):
        print("저는 데사스 {name} 입니다.".format(name=self.name))
                
    def coding(self):
        print("Jupyter Notebook 에서 pandas 라이브러리를 이용해서 개발을 합니다.")
```

```python
class Human():
    
    def __init__(self, name):
        self.name = name
        self.age = 0
        print("{name} 이라는 사람이 태어났다.".format(name = self.name))
        
    def after_a_year(self):
        self.age += 1
        print("{name} 이라는 사람이 나이를 1살 더 먹었다.".format(name = self.name))
        
    def __str__(self):
#         return self.name]
        return "'{name}' 이라는 사람".format(name = self.name)

    def __del__(self):
        print("{name} 이라는 사람이 메모리에서 잊혀졌다.".format(name = self.name))
```

```python
human = Human("이성희")
```

```python
human.after_a_year()
```

```python
print(human)
```

```python
# 파이썬이 이 사람을 기억 못할 때 => 더이상 필요없다고 생각했을때 
# 메모리에서 지워버리는 겁니다.
del human
```
