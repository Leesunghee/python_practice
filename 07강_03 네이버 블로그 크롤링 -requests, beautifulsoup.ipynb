```python
import requests
```

```python
# import bs4
from bs4 import BeautifulSoup
```

```python
BASE_URL = "http://search.naver.com/search.naver?where=post&sm=tab_jum&query=h.e.r"
```

```python
# 1. requests
#    Crawling
#   Requests: HTTP for Humans

response = requests.get(BASE_URL)
# requests?
```

```python
response.status_code    # 2XX => Successful ( 200 OK )
                        # 3XX => Redirection
                        # 4XX => Failed ( 403 Forbidden )
```

```python
# Crawling => Crawling, Parsing ( Crawling <<< Parsing )
dom = BeautifulSoup(response.content, "html.parser")
# dom => Document Object Model 
```

```python
BeautifulSoup?
```

```python
dom
```

```python
# Parsing

# tag => "" ( "div", "a" )
# id =>   unique   => "#"
# class =>   1~n =>    tag  => "."
```

```python
post_elements = dom.select("li.sh_blog_top")
```

```python
post_elements
```

```python
len(post_elements)
```

```python
post_element = post_elements[0]
post_element
```

```python
title_element = post_element.select_one("a.sh_blog_title")
title_element.text
```

```python
title_element.get("href")
```

```python
for post_element in post_elements:
    title_element = post_element.select_one("a.sh_blog_title")
    title = title_element.text
    url = title_element.get("href")
    print(title, url)
```

```python
#   
#  => DB
```

```python
import pandas as pd
df = pd.DataFrame(columns=["url", "title"])
df
```

```python
for post_element in post_elements:
    title_element = post_element.select_one("a.sh_blog_title")
    title = title_element.text
    url = title_element.get("href")
    df.loc[len(df)] = [url, title]
```

```python
df
```

```python
df.to_csv("her.csv")
```
